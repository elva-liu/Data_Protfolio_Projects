```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
GRU_simple = [383304.55, 355624.41, 486365]
GRU_enhanced= [388447.72, 357643.47, 532230]
GRU_enhanced_weather= [277898.99, 260747.72, 289833.28]

months = [12, 24, 36]

LSTM_simple = [477379.28, 425773.76, 568542.8]
LSTM_enhanced= [489990.41, 631796.88, 906004.8]
LSTM_enhanced_weather= [724795, 712680.71, 714709.94]
```

```python
# Create the plot
fig, ax1 = plt.subplots(figsize=(6,4),facecolor='lightgray')

ax1.plot(months, GRU_simple,'b-', marker="+", label='GRU_simple')
ax1.plot(months, GRU_enhanced,'r-', marker="+", label='GRU_enhanced')
ax1.plot(months, GRU_enhanced_weather,'g-', marker="+", label='GRU_enhanced_weather')
ax1.set_xlabel('Months')
ax1.set_ylabel('Mean Absolute Error', color='b')
ax1.tick_params(axis='y', labelcolor='b')
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
# Add titles and adjust layout
plt.title("Simple | Enhanced | Enhanced + Weather GRU [12-24-36] Months",fontsize=10, pad=15)
fig.tight_layout()

# Add legends
ax1.legend(loc='upper left')

fig=plt.gcf()
fig.savefig("pics\\energy_9.png", dpi=300)
# Display the plot
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(6,4),facecolor='lightblue')

ax1.plot(months, LSTM_simple,'b-', marker="+", label='LSTM_simple')
ax1.plot(months, LSTM_enhanced,'r-', marker="+", label='LSTM_enhanced')
ax1.plot(months, LSTM_enhanced_weather,'g-', marker="+", label='LSTM_enhanced_weather')
ax1.set_xlabel('Months')
ax1.set_ylabel('Mean Absolute Error', color='b')
ax1.tick_params(axis='y', labelcolor='b')
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
plt.title("Simple | Enhanced | Enhanced + Weather LSTM [12-24-36] Months",fontsize=10, pad=15)
fig.tight_layout()

ax1.legend(loc='upper left')
fig=plt.gcf()
fig.savefig("pics\\energy_10.png", dpi=300)
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(7,4),facecolor='yellow')

ax1.plot(months, GRU_enhanced_weather,'b-', marker="+", label='GRU_enhanced_weather')
ax1.plot(months, LSTM_simple,'r-', marker="+", label='LSTM_simple')

ax1.set_xlabel('Months')
ax1.set_ylabel('Mean Absolute Error', color='b')
ax1.tick_params(axis='y', labelcolor='b')
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
plt.title("Best Performance of each GRU  vs LSTM [12-24-36] Months",fontsize=15, pad=15)
fig.tight_layout()

ax1.legend(loc='upper left')

plt.show()
```

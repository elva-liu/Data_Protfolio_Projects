```python
import numpy as np
import matplotlib.pyplot as plt

def linear_to_stochastic(linear_data, noise_std):
    """
    Converts linear data to stochastic data by adding Gaussian noise.

    Args:
        linear_data (numpy.ndarray): The linear data.
        noise_std (float): The standard deviation of the Gaussian noise.

    Returns:
        numpy.ndarray: The stochastic data.
    """
    noise = np.random.normal(0, noise_std, len(linear_data))
    stochastic_data = linear_data + noise
    return stochastic_data

# Generate linear data
time = np.arange(0, 10, 0.1)
linear_data = 2 * time + 5  # Example: y = 2x + 5

# Add stochasticity
noise_std = 5  # Adjust for more/less noise
stochastic_data = linear_to_stochastic(linear_data, noise_std)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(time, linear_data, label='Linear Growth')
plt.plot(time, stochastic_data, label='Stochastic Growth', alpha=0.7)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Linear vs. Stochastic Growth')
plt.legend()
plt.grid(True)
plt.show()

# Example with a different noise level.
noise_std2 = 1
stochastic_data2 = linear_to_stochastic(linear_data, noise_std2)

plt.figure(figsize=(10, 6))
plt.plot(time, linear_data, label='Linear Growth')
plt.plot(time, stochastic_data2, label='Stochastic Growth (low noise)', alpha=0.7)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Linear vs. Stochastic Growth (Low Noise)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.arange(0, 60, 1)
time
```

```python
def linear_to_stochastic(linear_data, noise_std):
    noise = np.random.normal(0, noise_std, len(linear_data))
    stochastic_data = linear_data + noise
    return stochastic_data
```

```python
cumulative_increases = np.linspace(333501, 333501*12*5, 50)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(cumulative_increases, label='Linear Growth')
#plt.plot(time, stochastic_data, label='Stochastic Growth', alpha=0.7)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Linear vs. Stochastic Growth')
plt.legend()
plt.grid(True)
plt.show()
```

```python
noise_std = 333501*2
linear_to_stochastic(cumulative_increases, noise_std)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(cumulative_increases, label='Linear Growth')
plt.plot(linear_to_stochastic(cumulative_increases, noise_std), label='Stochastic Growth', alpha=0.7)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Linear vs. Stochastic Growth')
plt.legend()
plt.grid(True)
plt.show()
```
